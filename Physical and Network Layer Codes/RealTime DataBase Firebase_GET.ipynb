```python
import time
import csv

# Step 3: Fetch Data from Firebase
def fetch_data():
    # Get data from the "Smart_Energie" path
    ref = db.reference('/Smart_Energie')
    data = ref.get()
    return data

# Step 4: Save Data to CSV
def save_to_csv(data):
    # Path where the CSV file will be saved in Google Drive
    file_path = '/content/drive/My Drive/firebase_data.csv'

    # Prepare the data for CSV format
    records = [{
        'Current': data['data'].get('Current'),
        'Voltage': data['data'].get('Voltage'),
        'kWh': data['data'].get('kWh'),
        'Temps': data.get('temps')
    }]

    # Write to CSV
    with open(file_path, mode='a', newline='') as file:  # 'a' to append new data
        writer = csv.DictWriter(file, fieldnames=['Current', 'Voltage', 'kWh', 'Temps'])
        # If the file is empty, write the header row
        if file.tell() == 0:
            writer.writeheader()
        writer.writerows(records)

    print(f"Data written to {file_path}")

# Step 5: Run every 10 seconds
while True:
    data = fetch_data()
    if data:
        save_to_csv(data)
    time.sleep(10)
```
